```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://en.wikipedia.org/wiki/Main_Page")
page
```

```python
page.content#to see content
```

```python
soup=BeautifulSoup(page.content)# to see all the code of the content
soup
```

```python
print(soup.prettify())# to display it in particular format
```

```python
header_tags=soup.find_all(['h1','h2','h3','h4','h5','h6'])
header_tags
```

```python
header=[]
for i in header_tags:
    header.append(i.text.replace("\n",""))
header
```

```python
import pandas as pd
headers=pd.DataFrame({})
headers['header_tags']=header
headers
```
