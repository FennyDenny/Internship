```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.imdb.com/list/ls055592025/")
page
```

```python
soup=BeautifulSoup(page.content)
films=soup.find_all(class_='lister-item-header')
films
```

```python
first_films=[]
for i in films:
    first_films.append(i.text.replace("\n",""))
first_films
```

```python
import pandas as pd
films=pd.DataFrame({})
films['first_film']=first_films
films
```
