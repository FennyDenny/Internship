```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page=requests.get('https://bookpage.com/reviews')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
book_names=soup.find_all(class_='italic')
book_names
```

```python
booknames=[]
for i in book_names:
    booknames.append(i.text.replace("\n","").replace("★","").strip())
    
booknames
```

```python
author_names=soup.find_all(class_="sans bold")
author_names
```

```python
authornames=[]
for i in author_names:
    authornames.append(i.text.replace("\n",""))
authornames
```

```python
genres=soup.find_all(class_="genre-links hidden-phone")
genres
```

```python
genres_name=[]
for i in genres:
    genres_name.append(i.text.replace("\n",""))
genres_name
```

```python
reviews=soup.find_all(class_="excerpt")
reviews
```

```python
review_content=[]
for i in reviews:
    review_content.append(i.text.replace("\n",""))
review_content
```

```python
Books_review=pd.DataFrame({})
Books_review['book_names']=booknames
Books_review['author_names']=authornames
Books_review['genres_names']=genres_name
Books_review['reviews']=review_content
Five_Books=Books_review.iloc[0:5,:]
Five_Books
```
