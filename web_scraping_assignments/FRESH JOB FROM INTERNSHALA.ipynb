```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page=requests.get("https://internshala.com/fresher-jobs")
page
```

```python
soup=BeautifulSoup(page.content)
jobtitle=soup.find_all('div',class_="heading_4_5 profile")
jobtitle
```

```python
company=soup.find_all('a',class_="link_display_like_text")
company
```

```python
ctc=soup.find_all('div',class_="item_body")
```

```python
#applydate=soup.find_all(class_="item_body")
#applydate
```

```python
job_title=[]
for i in jobtitle:
    job_title.append(i.text.replace("\n","").strip())
job_title
    
```

```python
company_name=[]
for i in company:
    company_name.append(i.text.replace("\n","").strip())
print(company_name)
    
```

```python
ctc_and_salary=[]

j=0
for i in ctc:
    if (j%3!=0):
        ctc_and_salary.append(i.text.replace("\n","").strip())
    j=j+1
    
ctc_salary=[]
apply_date=[]
j=0
for i in ctc_and_salary:
    if (j%2!=0):
        apply_date.append(i)
    else:
        ctc_salary.append(i)
    j=j+1
print(ctc_salary)
print(apply_date)

    
```

```python
fresher_jobs=pd.DataFrame({})
fresher_jobs['jobtitle']=job_title
fresher_jobs['company_name']=company_name
fresher_jobs['ctc']=ctc_salary
fresher_jobs['apply_date']=apply_date
fresher_jobs
```
