```python
import selenium 
import pandas as pd
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
import re
```

```python
driver=webdriver.Chrome("chromedriver.exe")
driver.maximize_window()
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/")
```

```python
search_video_element=driver.find_element_by_id("searchInput")
search_video_element.send_keys("most viewed videos on YouTube")
```

```python
click_element=driver.find_element_by_id("searchButton")
click_element.click()
```

```python
rank_elements=driver.find_elements_by_xpath("//div[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[1]")
rank_elements
```

```python
rank_list=[]
for elem in rank_elements:
    rank=elem.text
    rank_list.append(rank)
rank_list
```

```python
name_elements=driver.find_elements_by_xpath("//div[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[2]")
name_elements
```

```python
name_list=[]
for elem in name_elements:
    #name=elem.text
    name =re.sub(r"(\")|(\[.*?\])", "", elem.text)
    name_list.append(name)
name_list
```

```python
artists_elements=driver.find_elements_by_xpath("//div[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[3]")
artists_elements
```

```python
artists_list=[]
for elem in artists_elements:
    artists=elem.text
    artists_list.append(artists)
artists_list
```

```python
date_elements=driver.find_elements_by_xpath("//div[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[5]")
date_elements
```

```python
date_list=[]
for elem in date_elements:
    date=elem.text
    date_list.append(date)
date_list
```

```python
view_elements=driver.find_elements_by_xpath("//div[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[4]")
view_elements
```

```python
view_list=[]
for elem in view_elements:
    view=elem.text
    view_list.append(view)
view_list
```

```python
import pandas as pd

youtube_video_df=pd.DataFrame({})
youtube_video_df["Rank"]=rank_list
youtube_video_df["Name"]=name_list
youtube_video_df["Artist"]=artists_list
youtube_video_df["Upload date"]=date_list
youtube_video_df["Views in billions"]=view_list

youtube_video_df
```
