```python
import selenium 
import pandas as pd
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
driver=webdriver.Chrome("chromedriver.exe")
driver.maximize_window()
```

```python
driver.get("https://www.imdb.com/list/ls095964455")
```

```python
group_box_elements=trending_element=driver.find_elements_by_xpath("//div[@id='main']/div/div[3]/div[3]/div")
group_box_elements
```

```python
name_list=[]
year_list=[]
genre_list=[]
rating_list=[]
vote_list=[]
time_list=[]
for elem in group_box_elements:
    #fetchning name elements
    try:
        name_element=elem.find_element_by_xpath(".//div[@class='lister-item-content']/h3/a")
        name=name_element.text
        name_list.append(name)
    except:
        name_list.append('-')
     #fetchning year elements
    try:
        year_element=elem.find_element_by_xpath(".//div[@class='lister-item-content']/h3/span[2]")
        year=year_element.text
        year_list.append(year)
    except:
        year_list.append('-')
     #fetchning genre elements
    try:
        genre_element=elem.find_element_by_xpath(".//span[@class='genre']")
        genre=genre_element.text
        genre_list.append(genre)
    except:
        genre_list.append('-')
      #fetchning run time elements
    try:
        time_element=elem.find_element_by_xpath(".//span[@class='runtime']")
        time=time_element.text
        time_list.append(time)
    except:
        time_list.append('-')  
        
     #fetchning rating elements
    try:
        rating_element=elem.find_element_by_xpath("./div[2]/div[1]/div[1]/span[2]")
        rating=rating_element.text
        rating_list.append(rating)
    except:
        rating_list.append('-')      
     #fetchning vote elements
    try:
        vote_element=elem.find_element_by_xpath(".//span[@name='nv']")
        vote=vote_element.text
        vote_list.append(vote)
    except:
        vote_list.append('-')      
    
        
```
