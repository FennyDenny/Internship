```python
import selenium 
import pandas as pd
from selenium import webdriver
import time
```

```python
driver=webdriver.Chrome("chromedriver.exe")
driver.maximize_window()
```

```python
driver.get("http://statisticstimes.com/")
```

```python
search_element=driver.find_element_by_xpath("//button[@class='dropbtn'][contains(text(),'Economy')]")
search_element.click()
time.sleep(3)
```

```python
menu_element=driver.find_element_by_xpath("//button[@class='dropbtn'][contains(text(),'Economy')]/../div/a[3]")
#menu_element.click()
url=menu_element.get_attribute('href')
driver.get(url)
time.sleep(3)
```

```python
driver.find_element_by_xpath("//a[contains(text(),'GDP of Indian states')]").click()
time.sleep(3)
```

```python
rank_elements=driver.find_elements_by_xpath("//table[@id='table_id']/tbody/tr/td[1]")
#rank_elements
rank_list=[]
for elem in rank_elements:
    rank=elem.text
    rank_list.append(rank)
rank_list
```

```python
state_elements=driver.find_elements_by_xpath("//table[@id='table_id']/tbody/tr/td[2]")
#state_elements
state_list=[]
for elem in state_elements:
    state=elem.text
    state_list.append(state)
state_list
```

```python
GSDP_1920_elements=driver.find_elements_by_xpath("//table[@id='table_id']/tbody/tr/td[3]")
GSDP_1920_list=[]
for elem in GSDP_1920_elements:
    GSDP_1920=elem.text
    GSDP_1920_list.append(GSDP_1920)
GSDP_1920_list
```

```python
GSDP_1819_elements=driver.find_elements_by_xpath("//table[@id='table_id']/tbody/tr/td[4]")
GSDP_1819_list=[]
for elem in GSDP_1819_elements:
    GSDP_1819=elem.text
    GSDP_1819_list.append(GSDP_1819)
GSDP_1819_list
```

```python
Share_1819_elements=driver.find_elements_by_xpath("//table[@id='table_id']/tbody/tr/td[5]")
Share_1819_list=[]
for elem in Share_1819_elements:
    Share_1819=elem.text
    Share_1819_list.append(Share_1819)
Share_1819_list
```

```python
GDP_billion_elements=driver.find_elements_by_xpath("//table[@id='table_id']/tbody/tr/td[6]")
GDP_billion_list=[]
for elem in GDP_billion_elements:
    GDP_billion=elem.text
    GDP_billion_list.append(GDP_billion)
GDP_billion_list
```

```python
import pandas as pd
gdp_india_df=pd.DataFrame({})
gdp_india_df["Rank"]=rank_list
gdp_india_df["State"]=state_list
gdp_india_df["GSDP at current price (19-20)"]=GSDP_1920_list
gdp_india_df["GSDP at current price (18-19)"]=GSDP_1819_list
gdp_india_df["Share(18-19)"]=Share_1819_list
gdp_india_df["GDP($ billion)"]=GDP_billion_list
gdp_india_df
```
