```python
import selenium 
import pandas as pd
from selenium import webdriver
import time
```

```python
driver=webdriver.Chrome("chromedriver.exe")
driver.maximize_window()
```

```python
driver.get("https://www.billboard.com/")
```

```python
menu_element=driver.find_element_by_xpath("//div[@id='root']/div[2]/div[2]/header/div/ul/li[1]/a")
menu_element.click()
time.sleep(3)
```

```python
view_element=driver.find_element_by_xpath("//main[@id='main']/div[2]/div/div[1]/a/div[2]/div[2]")
view_element.click()
time.sleep(3)
```

```python
name_elements=driver.find_elements_by_xpath("//span[@class='chart-element__information__song text--truncate color--primary']")
name_elements
```

```python
name_list=[]
for elem in name_elements:
    name=elem.text
    name_list.append(name)
name_list
```

```python
len(name_list)
```

```python
artists_elements=driver.find_elements_by_xpath("//span[@class='chart-element__information__artist text--truncate color--secondary']")
artists_elements
```

```python
artists_list=[]
for elem in artists_elements:
    artists=elem.text
    artists_list.append(artists)
artists_list
```

```python
len(artists_list)
```

```python
last_week_elements=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--last']")
last_week_elements
```

```python
last_week_list=[]
for elem in last_week_elements:
    last_week=elem.text
    last_week_list.append(last_week)
last_week_list
```

```python
len(last_week_list)
```

```python
peak_rank_elements=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--peak']")
peak_rank_elements
```

```python
peak_rank_list=[]
for elem in peak_rank_elements:
    peak_rank=elem.text
    peak_rank_list.append(peak_rank)
peak_rank_list
```

```python
weeks_elements=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--week']")
weeks_elements
```

```python
weeks_on_board_list=[]
for elem in weeks_elements:
    weeks=elem.text
    weeks_on_board_list.append(weeks)
weeks_on_board_list
```

```python
len(weeks_on_board_list)
```

```python
import pandas as pd
billboard_df=pd.DataFrame({})
billboard_df["Song name"]=name_list
billboard_df["Artist name"]=artists_list
billboard_df["Last week rank"]=last_week_list
billboard_df["Peak rank"]=peak_rank_list
billboard_df["Weeks on board"]=weeks_on_board_list
billboard_df
```
