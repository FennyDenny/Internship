```python
import selenium 
import pandas as pd
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
driver=webdriver.Chrome("chromedriver.exe")
driver.maximize_window()
```

```python
driver.get("https://github.com/")
```

```python
menu_element=driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary")
menu_element.click()
```

```python
trending_element=driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a")
#menu_element.click()
url=trending_element.get_attribute('href')
driver.get(url)
time.sleep(3)
```

```python
group_box_elements=trending_element=driver.find_elements_by_xpath("//article[@class='Box-row']")
group_box_elements
```

```python
repository_title_elements=driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']")
repository_title_elements
```

```python
description_list=[]
repository_list=[]
contribution_count_list=[]
languages_list=[]
for elem in group_box_elements:
    #fetching repository title
    try:
        repository_title_element=elem.find_element_by_xpath("./h1[@class='h3 lh-condensed']/a")
        repository=repository_title_element.text
        repository_list.append(repository)
    except:
        repository_list.append('-')
        
    #fetching Description   
    try:
        description_element=elem.find_element_by_xpath("./p[@class='col-9 color-fg-muted my-1 pr-4']")
        description=description_element.text
        description_list.append(description)
    except:
        description_list.append('-')
    
    #fetching Contribution Count   
    try:
        contribution_elements=elem.find_elements_by_xpath(".//a[@class='d-inline-block']")
        contribution_count=len(contribution_elements)
        contribution_count_list.append(contribution_count)
    except:
        contribution_count_list.append('-')
    
     #fetching Langauges Used   
    try:
        languages_element=elem.find_element_by_xpath(".//span[@itemprop='programmingLanguage']")
        languages=languages_element.text
        languages_list.append(languages)
    except:
        languages_list.append('-')
        
        

        
        
    
```

```python
repository_list,description_list,contribution_count_list,languages_list
```

```python
import pandas as pd
github_df=pd.DataFrame({})
github_df["Repository title"]=repository_list
github_df["Repository description"]=description_list
github_df["Contributors count"]=contribution_count_list
github_df["Language used"]=languages_list

github_df
```
