```python
import selenium 
import pandas as pd
from selenium import webdriver
import time
```

```python
driver=webdriver.Chrome("chromedriver.exe")
driver.maximize_window()
```

```python
driver.get("https://www.naukri.com/")
```

```python
menu_element=driver.find_element_by_xpath("//a[@title='Search Recruiters']")
url=menu_element.get_attribute('href')
driver.get(url)
time.sleep(3)
```

```python
search_element=driver.find_element_by_xpath("//input[@name='qp']")
search_element.send_keys("Data science")
```

```python
click_element=driver.find_element_by_xpath("//button[@class='fl qsbSrch blueBtn']")
click_element.click()
```

```python
name_elements=driver.find_elements_by_xpath("//span[@class='fl ellipsis']")
name_elements
```

```python
name_list=[]
for elem in name_elements:
    name=elem.text
    name_list.append(name)
name_list
```

```python
len(name_list)
```

```python
designation_elements=driver.find_elements_by_xpath("//span[@class='ellipsis clr']")
designation_elements                                                  
```

```python
designation_list=[]
for elem in designation_elements:
    designation=elem.text
    designation_list.append(designation)
designation_list
```

```python
len(designation_list)
```

```python
skill_elements=driver.find_elements_by_xpath("//div[@class='hireSec highlightable']")
skill_elements
```

```python
skill_list=[]
for elem in skill_elements:
    skill=elem.text
    skill_list.append(skill)
skill_list
```

```python
len(skill_list)
```

```python
group_box_elements=driver.find_elements_by_xpath("//div[@class='recInfo']")
group_box_elements
```

```python
len(group_box_elements)
```

```python
location_list=[]
company_list=[]
for elem in group_box_elements:
    #fetching location elements
    try:
        location_element=elem.find_element_by_xpath(".//small[@class='ellipsis']")
        location=location_element.text
        location_list.append(location)
    except:
        location_list.append('-')
    #fetching company elements
    try:
        company_element=elem.find_element_by_xpath(".//a[2]")
        company=company_element.text
        company_list.append(company)
    except:
        company_list.append('-')
```

```python
location_list
```

```python
len(location_list)
```

```python
company_list
```

```python
len(company_list)
```

```python
import pandas as pd
naukari_df=pd.DataFrame({})
naukari_df["Name"]=name_list
naukari_df["Designation"]=designation_list
naukari_df["Company"]=company_list
naukari_df["Skills"]=skill_list
naukari_df["Location"]=location_list
naukari_df
```
