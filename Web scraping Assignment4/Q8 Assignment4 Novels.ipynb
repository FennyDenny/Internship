```python
import selenium 
import pandas as pd
from selenium import webdriver
import time
```

```python
driver=webdriver.Chrome("chromedriver.exe")
driver.maximize_window()
```

```python
driver.get("https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare/")
```

```python
title_elements=driver.find_elements_by_xpath("//div[@id='article-body-blocks']/div/table/tbody/tr/td[2]")
title_elements
```

```python
title_list=[]
for elem in title_elements:
    title=elem.text
    title_list.append(title)
title_list
```

```python
len(title_list)
```

```python
author_elements=driver.find_elements_by_xpath("//div[@id='article-body-blocks']/div/table/tbody/tr/td[3]")
author_elements
```

```python
author_list=[]
for elem in author_elements:
    author=elem.text
    author_list.append(author)
author_list
```

```python
len(author_list)
```

```python
volume_elements=driver.find_elements_by_xpath("//div[@id='article-body-blocks']/div/table/tbody/tr/td[4]")
volume_elements
```

```python
volume_list=[]
for elem in volume_elements:
    volume=elem.text
    volume_list.append(volume)
volume_list
```

```python
len(volume_list)
```

```python
publisher_elements=driver.find_elements_by_xpath("//div[@id='article-body-blocks']/div/table/tbody/tr/td[5]")
publisher_elements
```

```python
publisher_list=[]
for elem in publisher_elements:
    publisher=elem.text
    publisher_list.append(publisher)
publisher_list
```

```python
genre_elements=driver.find_elements_by_xpath("//div[@id='article-body-blocks']/div/table/tbody/tr/td[6]")
genre_elements
```

```python
genre_list=[]
for elem in genre_elements:
    genre=elem.text
    genre_list.append(genre)
genre_list
```

```python
len(genre_list)
```

```python
import pandas as pd
novels_df=pd.DataFrame({})
novels_df["Book name"]=title_list
novels_df["Author name"]=author_list
novels_df["Volumes sold"]=volume_list
novels_df["Publisher"]=publisher_list
novels_df["Genre"]=genre_list
novels_df
```
