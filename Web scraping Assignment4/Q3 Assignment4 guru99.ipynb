```python
import selenium 
import pandas as pd
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
driver=webdriver.Chrome("chromedriver.exe")
driver.maximize_window()
```

```python
driver.get("https://www.guru99.com/")
time.sleep(3)  
```

```python
search_element=driver.find_element_by_xpath("//input[@class='gsc-input']")
search_element.send_keys("selenium exception handling")
```

```python
search_button=driver.find_element_by_xpath("//button[@class='gsc-search-button gsc-search-button-v2']")
search_button.click()
time.sleep(3)       
```

```python
searched_element=driver.find_element_by_partial_link_text("Selenium Exception Handling")
url=searched_element.get_attribute('href')
url                                        
```

```python
driver.get(url)
name_elements=driver.find_elements_by_xpath("//table[@class='table table-striped']/tbody/tr/td[1]")
name_elements
```

```python
description_elements=driver.find_elements_by_xpath("//table[@class='table table-striped']/tbody/tr/td[2]")
description_elements
```

```python
name_list=[]

for elem in name_elements:
    name=elem.text
    name_list.append(name)

name_list
```

```python
description_list=[]

for elem in description_elements:
    description=elem.text
    description_list.append(description)

description_list
```

```python
import pandas as pd
selenium_exception_df=pd.DataFrame({})
selenium_exception_df["Name"]=name_list
selenium_exception_df["Description"]=description_list

selenium_exception_df
```
