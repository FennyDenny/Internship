```python
import selenium 
import pandas as pd
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
driver=webdriver.Chrome("chromedriver.exe")
```

```python
driver.get("https://www.myntra.com/shoes")
```

```python
filter_color_element=driver.find_element_by_xpath("//input[@value='Black']/following-sibling::div[@class='common-checkboxIndicator']")
filter_color_element.click()
time.sleep(3)
```

```python
filter_price_element=driver.find_element_by_xpath("//input[@value='6972.0 TO 13645.0']/following-sibling::div[@class='common-checkboxIndicator']")
filter_price_element.click()
time.sleep(3)
```

```python
urls=[]
first_page_link=driver.find_element_by_xpath("//li[@class='pagination-active']/a")
print(first_page_link.get_attribute('href'))
urls.append(first_page_link.get_attribute('href'))
```

```python
for i in driver.find_elements_by_xpath("//li[@class='pagination-number']/a"):
    urls.append(i.get_attribute('href'))
len(urls) 
```

```python
brand_list=[]
for i in urls[0:2]:
    driver.get(i)
    time.sleep(3)
    brand_elements=driver.find_elements_by_xpath("//h3[@class='product-brand']")
    for elem in brand_elements:
        brand=elem.text
        brand_list.append(brand)
brand_list
```

```python
len(brand_list)
```

```python
short_descr_list=[]
for i in urls[0:2]:
    driver.get(i)
    time.sleep(3)
    short_descr_elements=driver.find_elements_by_xpath("//h4[@class='product-product']")
    for elem in short_descr_elements:
        short_descr=elem.text
        short_descr_list.append(short_descr)
short_descr_list
```

```python
len(short_descr_list)
```

```python
price_list=[]
for i in urls[0:2]:
    driver.get(i)
    time.sleep(3)
    price_elements=driver.find_elements_by_xpath("//div[@class='product-price']")
    for elem in price_elements:
        price=elem.text
        price_list.append(price)
price_list
```

```python
len(price_list)
```

```python
import pandas as pd
shoes_df=pd.DataFrame({})
shoes_df["Brand"]=brand_list[0:100]
shoes_df["Description"]=short_descr_list[0:100]
shoes_df["Price"]=price_list[0:100]
shoes_df
```
