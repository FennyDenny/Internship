# Q2

```python
urls=[]
```

```python
for i in driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']"):
    urls.append(i.get_attribute('href'))
```

```python
try:
    next_button_element=driver.find_element_by_xpath("//li[@class='a-last']")
    next_button_element.click()
    time.sleep(3)
    for i in driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']"):
        urls.append(i.get_attribute('href'))
except:
    print("No 2nd page")
```

```python
try:
    next_button_element=driver.find_element_by_xpath("//li[@class='a-last']")
    next_button_element.click()
    time.sleep(3)
    for i in driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']"):
        urls.append(i.get_attribute('href'))
except:
    print("No 3rd page")   
```

```python
len(urls)
```

```python
brand_list=[]
name_list=[]
number_of_rating_list=[]
price_list=[]
delivery_list=[]
availability_list=[]
exchange_list=[]
rating=[]
for i in urls:
    driver.get(i)
    time.sleep(3)
    #Fetching the brand name
    try:
        brand_element=driver.find_element_by_xpath("//tr[@class='a-spacing-small'][1]/td[@class='a-span9']/span")
        brand=brand_element.text
        brand_list.append(brand)
    except:
        brand_list.append('-')
    
     #Fetching the name of the product
    try:
        name_element=driver.find_element_by_id("productTitle")
        name=name_element.text
        name_list.append(name)
    except:
        name_list.append('-')
     
    #Fetching the no. of ratings
    try:
        number_of_rating_element=driver.find_element_by_id("acrCustomerReviewText")
        number_of_rating=number_of_rating_element.text
        number_of_rating_list.append(number_of_rating)
    except:
        number_of_rating_list.append('-')
        
     #Fetching the price
    try:
        price_element=driver.find_element_by_id("priceblock_ourprice")
        price=price_element.text
        price_list.append(price)
    
    except:
        price_list.append('-')
        
    #Fetching the Expected delivery
    try:
        delivery_element=driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']/b")
        delivery=delivery_element.text
        delivery_list.append(delivery)
    except:
        delivery_list.append('-')
    
    #Fetching the Expected availability
    try:
        availability_element=driver.find_element_by_id("availability")
        availability=availability_element.text
        availability_list.append(availability)
    except:
        availability_list.append('-')
     
    #Fetching the Exchange/return
    
    try:
        exchange_element=driver.find_element_by_xpath("//a[@class='a-size-small a-link-normal a-text-normal']")
        exchange=exchange_element.text
        exchange_list.append(exchange)
    except:
        exchange_list.append('-')
    
    #Fetching the rating
    try:
        rating_element_link =driver.find_element_by_xpath("//a[@id='acrCustomerReviewLink'][1]")
        rating_element_link.click()
        time.sleep(3)
        
        rate_element=driver.find_element_by_xpath("//span[@data-hook='rating-out-of-text']")
        rate=rate_element.text
        rating.append(rate)
    except:
        rating.append('-')
    
        
    
    
    
```

```python
import pandas as pd
amazon_df=pd.DataFrame({})
amazon_df["Brand"]=brand_list
amazon_df["Product name"]=name_list
amazon_df["Number of rating"]=number_of_rating_list
amazon_df["Price"]=price_list
amazon_df["Delivery"]=delivery_list
amazon_df["Availability"]=availability_list
amazon_df["Return"]=exchange_list
amazon_df["Rating"]=rating
amazon_df["Brand url"]=urls
amazon_df
```
