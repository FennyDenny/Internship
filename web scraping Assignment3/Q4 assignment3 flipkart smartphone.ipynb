```python
import selenium 
import pandas as pd
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
driver=webdriver.Chrome("chromedriver.exe")
```

```python
driver.get("https://www.flipkart.com/")
```

```python
close_btn=driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']")
close_btn.click()
time.sleep(3)
```

```python
search_phone_element=driver.find_element_by_name("q")
search_phone_element.send_keys("samsung mobile")
```

```python
search_button_element=driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
search_button_element.click()
time.sleep(3)
```

```python
urls=[]
```

```python
for i in driver.find_elements_by_xpath("//a[@class='_1fQZEK']"):
    urls.append(i.get_attribute('href'))

```

```python
urls
```

```python
brand_list=[]
model_list=[]
color_list=[]
ram_list=[]
storage_list=[]
primary_list=[]
secondary_list=[]
display_size_list=[]
display_resolution_list=[]
processor_list=[]
processor_core_list=[]
battery_list=[]
price_list=[]
for i in urls:
    driver.get(i)
    time.sleep(3)
    more_btn=driver.find_element_by_xpath("//button[@class='_2KpZ6l _1FH0tX']")
    more_btn.click()
    time.sleep(3)
    #fetching the brand name
    try:
        brand_element=driver.find_element_by_xpath("//span[@class='B_NuCI']")
        brand=brand_element.text
        brand_list.append(brand)
    except:
        brand_list.append('-')
    #fetching the smartphone name
    try:
        model_element=driver.find_element_by_xpath("//td[text()='Model Name']/../td[2]")
        model=model_element.text
        model_list.append(model)
    except:
        model_list.append('-')
    #fetching the color
    try:
        color_element=driver.find_element_by_xpath("//td[text()='Color']/../td[2]")
        color=color_element.text
        color_list.append(color)
    except:
        color_list.append('_')


    #fetching the Ram
      
    try:
        ram_element=driver.find_element_by_xpath("//td[text()='RAM']/../td[2]")
        ram=ram_element.text
        ram_list.append(ram)
    except:
        ram_list.append('-')
    
    #fetching the storage
    try:
        storage_element=driver.find_element_by_xpath("//td[text()='Expandable Storage']/../td[2]")
        storage=storage_element.text
        storage_list.append(storage)
    except:
        storage_list.append('-')
    #fetching the primary camera
    try:
        primary_camera_element=driver.find_element_by_xpath("//td[text()='Primary Camera']/../td[2]")
        primary=primary_camera_element.text
        primary_list.append(primary)
    except:
        primary_list.append('-')
    
    #fetching the secondary camera
    try:
        secondary_camera_element=driver.find_element_by_xpath("//td[text()='Secondary Camera']/../td[2]")
        secondary=secondary_camera_element.text
        secondary_list.append(secondary)
    except:
        secondary_list.append('-')
    #fetching the display
    try:
        display_size_element=driver.find_element_by_xpath("//td[text()='Display Size']/../td[2]")
        display=display_size_element.text
        display_size_list.append(display)
    except:
        display_size_list.append('-')
     #fetching the display resolution
    try:
        display_resolution_element=driver.find_element_by_xpath("//td[text()='Resolution']/../td[2]")
        display_resolution=display_resolution_element.text
        display_resolution_list.append(display_resolution)
    except:
        display_size_list.append('-')
     #fetching the display processor
    try:
        processor_element=driver.find_element_by_xpath("//td[text()='Processor Type']/../td[2]")
        processor=processor_element.text
        processor_list.append(processor)
    except:
        processor_list.append('-')
    
     #fetching the processor crore
    try:
        processor_core_element=driver.find_element_by_xpath("//td[text()='Processor Core']/../td[2]")
        processor_core=processor_core_element.text
        processor_core_list.append(processor_core)
    except:
        processor_core_list.append('-')
    #fetching the processor crore
    
    try:
        battery_element=driver.find_element_by_xpath("//td[text()='Battery Capacity']/../td[2]")
        battery=battery_element.text
        battery_list.append(battery)
    except:
        battery_list.append('-')
        
    #Fetching price
    try:
        price_tags=driver.find_element_by_xpath("//div[@class='_30jeq3 _16Jk6d']")
        price=price_tags.text
        price_list.append(price)
    except:
        price_list.append('-')
 
    
    
    
    
```

```python
import pandas as pd
smartphone=pd.DataFrame({})
smartphone["Brand"]=brand_list
smartphone["Model"]=model_list
smartphone["Price"]=price_list
smartphone["Color"]=color_list
smartphone["RAM"]=ram_list
smartphone["Storage"]=storage_list
smartphone["Primary Camera"]=primary_list
smartphone["Secondary Camera"]=secondary_list
smartphone["Display Size"]=display_size_list
smartphone["Resolution"]=display_resolution_list
smartphone["Processor"]=processor_list
smartphone["Processor Core"]=processor_core_list
smartphone["Battery"]=battery_list
smartphone["URL"]=urls
smartphone
```
