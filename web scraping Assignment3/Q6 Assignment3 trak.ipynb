```python
import selenium 
import pandas as pd
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
driver=webdriver.Chrome("chromedriver.exe")
driver.maximize_window()
```

```python
driver.get("https://www.trak.in/")
```

```python
funding_deal_link_element= driver.find_element_by_link_text("Funding Deals")
funding_deal_link_element.click()
time.sleep(3)
```

# Getting July 2020 Data

```python
july_2020_table_element=driver.find_element_by_xpath("//h2[text()='July, 2020']/following-sibling::div[1]//div[@class='dataTables_scrollBody']/table")
july_2020_table_element
```

```python
july_2020_table_rows_elements=july_2020_table_element.find_elements_by_xpath("./tbody/tr")
len(july_2020_table_rows_elements)
```

```python
july_2020_sr_no_list=[]
july_2020_date_list=[]
july_2020_startup_name_list=[]
july_2020_vertical_list=[]
july_2020_sub_vertical_list=[]
july_2020_city_list=[]
july_2020_investor_name_list=[]
july_2020_investor_type_list=[]
july_2020_ammount_list=[]
for row_elem in july_2020_table_rows_elements:
    july_2020_sr_no_list.append(row_elem.find_element_by_xpath("./td[@class='column-1']").text)
    july_2020_date_list.append(row_elem.find_element_by_xpath("./td[@class='column-2']").text)
    july_2020_startup_name_list.append(row_elem.find_element_by_xpath("./td[@class='column-3']").text)
    july_2020_vertical_list.append(row_elem.find_element_by_xpath("./td[@class='column-4']").text)
    july_2020_sub_vertical_list.append(row_elem.find_element_by_xpath("./td[@class='column-5']").text)
    july_2020_city_list.append(row_elem.find_element_by_xpath("./td[@class='column-6']").text)
    july_2020_investor_name_list.append(row_elem.find_element_by_xpath("./td[@class='column-7']").text)
    july_2020_investor_type_list.append(row_elem.find_element_by_xpath("./td[@class='column-8']").text)
    july_2020_ammount_list.append(row_elem.find_element_by_xpath("./td[@class='column-9']").text)
    
```

```python
july_2020_sr_no_list,july_2020_date_list,july_2020_startup_name_list,july_2020_vertical_list,july_2020_sub_vertical_list,july_2020_city_list,july_2020_investor_name_list,july_2020_investor_type_list,july_2020_ammount_list
```

```python
sep_2020_sr_no_list,sep_2020_date_list,sep_2020_startup_name_list,sep_2020_vertical_list,sep_2020_sub_vertical_list,sep_2020_city_list,sep_2020_investor_name_list,sep_2020_investor_type_list,sep_2020_ammount_list
```

```python
sr_no_list=july_2020_sr_no_list + aug_2020_sr_no_list + sep_2020_sr_no_list
date_list=july_2020_date_list + aug_2020_date_list + sep_2020_date_list
startup_name_list= july_2020_startup_name_list + aug_2020_startup_name_list + sep_2020_startup_name_list
vertical_list=july_2020_vertical_list + aug_2020_vertical_list + sep_2020_vertical_list
sub_vertical_list=july_2020_sub_vertical_list + aug_2020_sub_vertical_list + sep_2020_sub_vertical_list
city_list=july_2020_city_list + aug_2020_city_list + sep_2020_city_list
investor_name_list=july_2020_investor_name_list + aug_2020_investor_name_list + sep_2020_investor_name_list
investor_type_list=july_2020_investor_type_list + aug_2020_investor_type_list + sep_2020_investor_type_list
ammount_list=july_2020_ammount_list + aug_2020_ammount_list + sep_2020_ammount_list
```
